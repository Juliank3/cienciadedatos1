```python
#Generando el Data Frame con Pandas
import pandas as pd

data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}

df1 = pd.DataFrame(data)

df1
```

```python
#Generar la Columna1 dónde se muestre el % del gasto con las ventas
df1['% Gastos'] = (df1['Gastos']/df1['Ventas'])*100
df1['Meta'] = df1['Ventas'].apply(lambda x: 'Meta Superada' if x >= 30000 else 'Meta no Superada')
df1['Sobre Gastos'] = df1.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else "Sin sobre Gastos", axis=1)
df1
```

```python
total_ult_3_meses = df1['Ventas'][-3:].sum()
print("Total de Ventas de los últimos 3 meses:", total_ult_3_meses)
```

```python
# Crearemos un diccionario para asignar un valor numérico a cada mes, luego utilizaremos .map() para volver los nombres en valor númerico,
# y que de esta manera podamos utilizar los meses entre febrero y Abril.
meses_numericos = {
    'Enero': 1,
    'Febrero': 2,
    'Marzo': 3,
    'Abril': 4,
    'Mayo': 5
}

# Filtrar las filas del DataFrame para los meses de 'Febrero' a 'Abril'
gastos_feb_abr = df1[(df1['Mes'].map(meses_numericos) >= meses_numericos['Febrero']) & (df1['Mes'].map(meses_numericos) <= meses_numericos['Abril'])]

# Calcular el total de los 'Gastos' en esas filas
total_gastos_feb_abr = gastos_feb_abr['Gastos'].sum()

print("Total de Gastos de los meses de Febrero a Abril:", total_gastos_feb_abr)
```
