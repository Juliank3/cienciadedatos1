```python
# Crear un diccionario de ventas mensuales

ventas_mensuales = {
    "Cliente1": 5000,
    "Cliente2": 7500,
    "Cliente3": 3000,
    "Cliente4": 6000,
}
```

```python
ventas_mensuales
```

```python
#Código para adjuntar más clientes
nom_cliente = 'Cliente7'
import_cliente = 10000

ventas_mensuales[nom_cliente] = import_cliente
ventas_mensuales
```

```python
#Se generará el programa que obtenga cómo resultado una serie con el máximo, mínimo y promedio de las ventas mensuales

#Generaremos una lista a partir del diccionario ventas_mensuales
valores_ventas = list(ventas_mensuales.values())

max_valor = max(valores_ventas)
min_valor = min(valores_ventas)
promedio = sum(valores_ventas) / len(valores_ventas)

print(f"El valor máximo de ventas es: {max_valor}, el valor mínimo de ventas es: {min_valor} y el promedio de ventas es: {promedio}.")
```
