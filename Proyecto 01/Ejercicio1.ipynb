```python
#Ejecutando Librería Panda
import pandas as pd
```

```python
#Subiendo la Base de Datos de Empresa "ABC S.A.C." a Data Frames
df1 = pd.read_csv('/content/Trabajo Python.csv', sep=';') # sep=';'
df1.head()
```

```python
#Convirtiendo la columna emisión en datetime
df1['Emisión'] = pd.to_datetime(df1['Emisión'], format='%d/%m/%Y')
```

```python
#Ordenando de la fecha más antigua a la más reciente.
df2 = df1.sort_values(by='Emisión', ascending=False)
df2
```

```python
#Generando la venta de los últimos 5 días
fecha_inicio = '2022-05-7'
fecha_fin = '2022-05-11'

df_filtrado = df2[(df2['Emisión'] >= fecha_inicio) & (df2['Emisión'] <= fecha_fin)]
df_filtrado
```

```python
#Se creó la columna importe neto con dos decimales.
df_filtrado['Monto Descuento'] = df_filtrado['Importe'] * df_filtrado['Descuento']
df_filtrado['Monto Descuento'] = df_filtrado['Monto Descuento'].round(2)
df_filtrado['Importe Neto'] = df_filtrado['Importe'] - df_filtrado['Monto Descuento']
```
