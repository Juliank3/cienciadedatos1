```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def generar_mapa_de_calor(dataframe):
    # Seleccionar las columnas de interés
    columnas_interes = ["Millas_Por_Galon", "Peso_En_Libras", "Desplazamiento", "Aceleracion", "Caballo_De_Fuerza"]
    datos_interes = dataframe[columnas_interes]

    # Calcular la matriz de correlación
    matriz_correlacion = datos_interes.corr()

    # Crear un mapa de calor
    plt.figure(figsize=(10, 8))
    sns.heatmap(matriz_correlacion, annot=True, cmap="RdYlGn_r", fmt=".2f", linewidths=.5)
    plt.title("Mapa de Calor de Correlaciones")
    plt.show()

# Cargar el DataFrame desde el archivo "autos.csv"
df = pd.read_csv('/content/autos.csv', sep=';')

# Llamar a la función con el DataFrame proporcionado
generar_mapa_de_calor(df)
```

```python
def generar_mapa_de_calor_por_pais(dataframe, pais):
    # Filtrar el DataFrame para el país específico
    datos_pais = dataframe[dataframe['Pais_Origen'] == pais]

    # Seleccionar las columnas de interés
    columnas_interes = ["Millas_Por_Galon", "Peso_En_Libras", "Desplazamiento", "Aceleracion", "Caballo_De_Fuerza"]
    datos_interes = datos_pais[columnas_interes]

    # Calcular la matriz de correlación
    matriz_correlacion = datos_interes.corr()

    # Crear un mapa de calor
    plt.figure(figsize=(10, 8))
    sns.heatmap(matriz_correlacion, annot=True, cmap="RdYlGn_r", fmt=".2f", linewidths=.5)
    plt.title(f"Mapa de Calor de Correlaciones - {pais}")
    plt.show()

# Cargar el DataFrame desde el archivo "autos.csv"
df = pd.read_csv('/content/autos.csv', sep=';')

# Obtener la lista de países únicos en la columna "Pais_Origen"
paises_unicos = df['Pais_Origen'].unique()

# Generar un mapa de calor por cada país
for pais in paises_unicos:
    generar_mapa_de_calor_por_pais(df, pais)
```
