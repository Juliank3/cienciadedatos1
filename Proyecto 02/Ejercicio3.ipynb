```python
import pandas as pd
```

```python
df = pd.read_csv('/content/autos.csv', sep=';') # sep=';'
df.head()
```

```python
# Cambiar los nombres de las columnas
df.rename(columns={'Caballo_De_Fuerza': 'Caballo_Fuerza_Var2'}, inplace=True)

# Definir la función para asignar categorías
def asignar_categoria(valor, media, desviacion_estandar):
    if valor > media + desviacion_estandar:
        return 'Alto'
    elif media - desviacion_estandar <= valor <= media + desviacion_estandar:
        return 'Medio'
    else:
        return 'Bajo'

# Calcular la media y la desviación estándar de la columna 'Caballo_Fuerza_Var2'
media = df['Caballo_Fuerza_Var2'].mean()
desviacion_estandar = df['Caballo_Fuerza_Var2'].std()

# Aplicar la función para asignar categorías y crear la nueva columna 'Categoría'
df['Categoría'] = df['Caballo_Fuerza_Var2'].apply(lambda x: asignar_categoria(x, media, desviacion_estandar))
```
