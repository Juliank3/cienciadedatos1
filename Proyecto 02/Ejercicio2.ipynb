```python
import pandas as pd
```

```python
df = pd.read_csv('/content/autos.csv', sep=';') # sep=';'
df.head()
```

```python
# Vamos a tratar las variables cambiando su nombre por Var1 y Var2 para proceder con sus cálculos.
df.rename(columns={'Aceleracion': 'Aceleracion_Var1', 'Caballo_De_Fuerza': 'Caballo_Fuerza_Var2'}, inplace=True)
```

```python
import numpy as np
```

```python
# Calcular el log10(Aceleracion_Var1 / Caballo_Fuerza_Var2)
df['calculo1'] = np.log10(df['Aceleracion_Var1'] / df['Caballo_Fuerza_Var2'])
df['calculo1'] = df['calculo1'].round(1)

# Calcular la expresión Sqrt(var1)*exp(var2)/200
df['calculo2'] = (np.sqrt(df['Aceleracion_Var1']) * np.exp(df['Caballo_Fuerza_Var2']))/200
df['calculo2'] = df['calculo2'].round(1)

# Calcular el valor condicional para la columna 'calculo3'
df['calculo3'] = np.where(df['Aceleracion_Var1'] > df['Caballo_Fuerza_Var2'], 5, 3)
df['calculo3'] = df['calculo3'].round(2)

# Calcular el valor 1/logn(var1/var2)*100 en una nueva columna
df['calculo4'] = 1 / np.log(df['Aceleracion_Var1'] / df['Caballo_Fuerza_Var2']) * 100
df['calculo4'] = df['calculo4'].round(1)

# Calcular la expresión var2**2/var1
df['calculo5'] = df['Aceleracion_Var1']**2 / df['Caballo_Fuerza_Var2']
df['calculo5'] = df['calculo5'].round(1)

df
```
